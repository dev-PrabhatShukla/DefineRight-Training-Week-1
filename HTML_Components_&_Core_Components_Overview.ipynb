<h1><b>Layout of Dash Apps:</b></h1>

* Hierarchical tree of components.
* It has mainly two popular modules :
    - Dash HTML Component : dash.html
    - Dash Core Component : dash.dcc

Let's look at the basic HTML and some of their known elements :     

<b>HTML(Hyper Text Markup Language) : </b>The standard markup language for structuring the web page.

* <B>HTML Documents -> Web Browsers -> Multi-media Web Pages.</b>

    - <b>Section Headings</b> : h1, h2, h3, h4, h5
    - <b>Paragraph</b> : p
    - <b>Content Div</b> : div(It is used as a container to group other HTML elements as one section).
    - <b>Line Break</b> : br
    - <b>Anchor</b> : a (creates hyperlink to other webpages).

* <b> HTML Elements : </b>HTML tags define the HTML elements.



<h2><b>Dash HTML Components Module :</b></h2>

* Contains components for every HTML tag.
    - from dash import html

* Example :      
    - html.H1(children = 'My Dashboard')


<h3>Let's build a <b>"World Happiness Dashboard"</b></h3> webpage with the help of <B>HTML Components</B>

<H3><B>Below is the same code :</b></h3>

```python
from dash import Dash, html

app = Dash()
app.layout = html.Div([
    html.H1('World Happiness Dashboard'),
    html.P(['This dashboard shows the happiness score.',
            html.Br(),
            html.A('World Happiness Report Data Source', href='https://worldhappiness.report', target='_blank')])
])

if __name__=='__main__':
    app.run_server(debug=True)
```

<h2><b>Dash Core Components Modules :</b></h2>

* Higher level components that can be interactive.
* Written with Javascript, HTML, CSS, etc.
    * from dash import dcc
* Examples :     
    * **Dropdown** : dcc.Dropdown
    * **Radio items** : dcc.RadioItems
    * **Graphs** : dcc.Graph
    * **Text input boxes** : dcc.Input
    * **Sliders & Range Slider** : dcc.Slider & dcc.RangeSlider
    * **Dropdown** : dcc.Dropdown

Now, we are going to build an interactive <b>World Happiness Dashboard</b> by using **Dash Core Components (dcc)** module.

So for that we continuing following the process of previously build webpage followed by making it more interactive using Dropdown, Radio, Graph, & other core dash components.

---

We have a dataset file in csv format named <b>world_happiness.csv</b> so we call that particular file here and named it as <b>happiness</b> dataset.

<H3><B>Below is the same code :</b></h3>

```python
from dash import Dash, html, dcc
import pandas as pd
import plotly.express as px

happiness = pd.read_csv('world_happiness.csv')
# "C:\Users\Prabhat Shukla\Desktop\My_first_dashboard\
app = Dash()
app.layout = html.Div([
    html.H1('World Happiness Dashboard'),
    html.P(['This dashboard shows the happiness score.',
            html.Br(),
            html.A('World Happiness Report Data Source', href='https://worldhappiness.report', target='_blank')]),
    html.Br(),
    dcc.RadioItems(options=happiness['region'].unique(), value='North America'),
    html.Br(),
    dcc.Checklist(options=happiness['region'].unique(), value=['Western Europe']),
    html.Br(),
    dcc.Dropdown(options=happiness['country'].unique(), value='United States'),
    html.Br(),
    dcc.Graph(figure=px.line(happiness[happiness['country'] == 'United States'], x='year', y='happiness_score',
                             title='Happiness Score in the USA')),
])

if __name__=='__main__':
    app.run_server(debug=True)
```
