<b>Let's look at what exactly <u>Dash</u> is: </b>




* Python library for building web applications.
* Developed by the plotly company.
* Built on top of plotly, Flask, & react
      -> Flask allowed dash to create web based applications.
      -> react is javascript library used here for building user interfaces.

Just because the integration of these two libraries we got the power of team of web developer. We can built web based applications with just Python & Dash.

<b>Who is <u>Dash</u> actually for? </b>


* It's for Data Analysts/Scientists with basic knowledge of Python.
* Share data visualization insights with customers, executives, rest of companies.
* Different industries : Finance, Medical, Marketing & so on.



<b>Why <u>Dash</u>? </b>



* Compare to regular data visualization libraries such as Matplotlib, Seaborn :


1. Dash can helps you to turns simple data into a well stuctured Dahboard.
2. Dash can also create interactive Dashboards.
3. Higly customizable (change sizing, fonts, color of the text, look of the graph, & so on).
4. Due to its web applications nature the dash Dashboard can deployed to servers and then share easily with other.
5. Dash is an open source library so you can use it for free.















**Parts of <u>Dash</u> :**

There is two main parts of the <b>Dash</b> :    
 a. **<u>Layout</u>**: Layout contains different components and the determence how the dashboard as per looks.
 Other popular components you can do in dash include drop down menu, radio items & so on.

 <b>Dash</b> allows us to quickly building components with two modules :     


*   **DCC** : Dash Core Components
*   **HTML** : Dash HTML Components.

b. **<u>Interactive Features</u>** : Callback functions



<H3><b>Create your first Dashboard</b></H3>

To create your first Dashboard you just need to first install any code editor (VS Code, Pycharm, IntelliJ, etc.)

> Also install the python in your system and the set that version of python as intrepreter in your code editor.

> Create a folder named <b>My_first_Dashboard</b> and open that folder in your code editor and then install the Dash package in that folder.

    - pip install dash


Create a python file named <b>dash_example</b> inside the <b>My_first_Dashboard</b> folder and write the below code in that particular file.

```python
from dash import Dash, HTML   # Importing the Dash library
app = Dash()

app.layout = html.Div('Welcome to my first dashboard')   # It's shows html view on the web page.

if __name__=='__main__':
  app.run_server(debug=True) # While development the debug needs to be true but in products it needs to be false.
```
